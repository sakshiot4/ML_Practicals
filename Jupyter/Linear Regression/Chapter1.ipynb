First Machine Learning Code

*_______This is a simple regression model(y=mx+c) that learns y = 2 * x.________*

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
X = [[1], [2], [3], [4]]
y = [2, 4, 6, 8]
model.fit(X, y)

user_value = int(input("Enter a number to predict: "))

prediction = model.predict([[user_value]])
print(f"Prediction for input {user_value} is : ", prediction[0])
```

```python
import matplotlib.pyplot as plt

#regression line.
plt.plot(X, model.predict(X), color = "red", label = "Regression line") #X value is predicted here.

#scatter plot
plt.scatter(X, y, color = 'blue', label = 'Data points')

#add legend on the right.
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
plt.title("Linear Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.tight_layout() #keeps the layout clean when legend is outside.
plt.show()
```
