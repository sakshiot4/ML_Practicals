## Handling missing values

```python
#to count the null values.
print(df.isnull().sum()) 

#fill the data with the mean value.
df["score"] = df["score"].fillna(df['score'].mean()) 

#drop the rows with missing values.
data = df.dropna()
```

## Data Check

```python
print(df.shape)
df.info()
print(df.describe())
print("Missing values: \n", df.isnull().sum())
```
