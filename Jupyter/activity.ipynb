## Q. Write a program that predicts the prices of banana given a linear pattern of weight and price

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
weights = np.array([1, 2, 3, 4, 5]).reshape(-1, 1) #features.
#print(weights.shape)
prices = np.array([30,60, 90, 120, 150]) #labels.

#first plot the graph to see if the linear regression is working.
plt.scatter(weights, prices, color = "blue")
```

```python
#initialize the model.
model = LinearRegression()

#train the model.
model.fit(weights, prices)

#predict the model.
predict = model.predict([[6]])
print("Price of 6kg bananas is ", predict)
```

```python
#plot the graph.
plt.plot(weights, model.predict(weights), color = "yellow", label="regression lines")
plt.scatter(weights, prices, color = "blue", label="Data Points")

#add legend on the right.
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
plt.title("Linear Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.tight_layout() #keeps the layout clean when legend is outside.
plt.show()
```

## Activity 2: 

```python
model = LinearRegression()
hours = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
marks = np.array([5, 10, 15, 20, 25])

model.fit(hours, marks)

plt.scatter(hours, marks, color = "blue")

user_value = int(input("Enter a number to predict: "))

prediction = model.predict([[user_value]])
print(f"Prediction for input {user_value} is : ", prediction[0])

```

```python
#regression line.
plt.plot(hours, model.predict(hours), color = "red", label = "Regression line") #X value is predicted here.

#scatter plot
plt.scatter(hours, marks, color = 'blue', label = 'Data points')

#add legend on the right.
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
plt.title("Linear Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.tight_layout() #keeps the layout clean when legend is outside.
plt.show()
```
